## Scatter plots

Kode sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges']) menggunakan pustaka Seaborn untuk membuat scatter plot (plot pen散布 (san-pu) = sebar) yang memvisualisasikan hubungan antara Indeks Massa Tubuh (IMT) dan biaya asuransi dalam DataFrame insurance_data Anda.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

1. Import Seaborn (Diasumsikan): Diasumsikan Anda sudah mengimpor Seaborn dengan perintah import seaborn as sns.

2. Scatter Plot:

* sns.scatterplot adalah fungsi dari Seaborn yang digunakan untuk membuat scatter plot.

* x=insurance_data['bmi']: Ini menetapkan kolom IMT (jika ada) sebagai data untuk sumbu horizontal (x).

* y=insurance_data['charges']: Ini menetapkan kolom biaya (jika ada) sebagai data untuk sumbu vertikal (y).


Kode sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges']) mengembangkan konsep scatter plot yang Anda lihat sebelumnya, tetapi dengan fitur tambahan.

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Plot Regresi:

* sns.regplot adalah fungsi dari Seaborn yang khusus dirancang untuk plot regresi.

* Ia mewarisi fungsionalitas dari sns.scatterplot untuk membuat scatter plot antara IMT dan biaya (seperti sebelumnya).

* Selain itu, fungsi ini menghitung garis regresi yang paling sesuai dengan titik data, bertujuan untuk menangkap tren keseluruhan dalam hubungan antara IMT dan biaya.

## Color-coded scatter plots

Kode sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker']) membuat scatter plot yang diberi warna berdasarkan variabel kategoris, untuk menunjukkan potensi perbedaan antar kelompok.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Scatter Plot dengan Pewarna (Hue):

* sns.scatterplot adalah fungsi dasar untuk membuat scatter plot.

* x=insurance_data['bmi'] dan y=insurance_data['charges'] menentukan data untuk sumbu x dan y (sama seperti sebelumnya).

* hue=insurance_data['smoker'] memperkenalkan konsep baru. Ini menetapkan warna untuk setiap titik data berdasarkan nilai pada kolom 'smoker' (perokok). Hal ini memungkinkan Anda untuk melihat apakah ada tren atau perbedaan dalam hubungan antara IMT dan biaya tergantung pada apakah seseorang perokok atau bukan.



Kode sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data) mengembangkan konsep scatter plot berwarna yang Anda lihat sebelumnya, tetapi selangkah lebih maju dengan menambahkan garis regresi.

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

Plot Model Linear:

* sns.lmplot adalah fungsi dari Seaborn yang khusus dirancang untuk membuat visualisasi dengan garis regresi untuk data kategoris.

* x="bmi" dan y="charges" menentukan data untuk sumbu x dan y (mirip dengan scatter plot).

* hue="smoker" menetapkan warna untuk setiap titik data dan garis regresi berdasarkan nilai pada kolom 'smoker' (seperti scatter plot berwarna).

* data=insurance_data menentukan DataFrame yang berisi data untuk plot.

Kode sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges']) menggunakan pustaka Seaborn untuk membuat jenis plot sebaran khusus yang disebut "swarm plot" (plot sebaran lebah).

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

Plot Sebaran Lebah (Swarm Plot):

* sns.swarmplot adalah fungsi dari Seaborn yang dirancang untuk memvisualisasikan data kategoris.

* x=insurance_data['smoker'] menetapkan kolom 'smoker' (kategoris) sebagai data untuk sumbu x.

* y=insurance_data['charges'] menetapkan kolom biaya (numerik) sebagai data untuk sumbu y.
