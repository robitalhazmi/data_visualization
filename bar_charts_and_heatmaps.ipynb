## Set up the notebook

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

## Load the data

Kode yang diberikan berikut ini menunjukkan bagaimana cara membaca data dari file CSV (Comma-Separated Values) yang berisi informasi mengenai delay penerbangan menggunakan pustaka Pandas di Python:

```python
# Path of the file to read
flight_filepath = "/content/flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

1. flight_filepath = "/content/flight_delays.csv": Baris pertama mendefinisikan variabel flight_filepath dan menetapkan path ke file CSV yang ingin dibaca. Ganti "/content/flight_delays.csv" dengan path sebenarnya dari file CSV Anda yang berisi data delay penerbangan

2. flight_data = pd.read_csv(flight_filepath, index_col="Month"): Baris kedua menggunakan fungsi pd.read_csv dari pustaka Pandas untuk membaca data dari file CSV. Fungsi ini mengambil beberapa argumen:
    
* flight_filepath: Path ke file CSV yang ingin dibaca.
    
* index_col="Month": Menentukan kolom "Month" sebagai index dari DataFrame yang dihasilkan. Ini berarti baris dalam DataFrame akan diidentifikasi berdasarkan nilai pada kolom "Month".

Setelah dieksekusi, kode ini akan menyimpan data yang dibaca dari file CSV ke dalam variable flight_data sebagai DataFrame Pandas. DataFrame adalah struktur data tabular yang lazim digunakan untuk analisis data.

## Examine the data

Baris kode ini menampilkan isi dari DataFrame flight_data ke console.

## Bar chart

Kode yang diberikan berikut ini akan menghasilkan visualisasi berupa bar chart yang menunjukkan rata-rata delay kedatangan bulanan untuk penerbangan Spirit Airlines:

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

1. Mengatur Ukuran dan Judul (sama seperti sebelumnya):

Baris-baris awal mengatur ukuran plot menjadi 10 x 6 inch dan menambahkan judul "Average Arrival Delay for Spirit Airlines Flights, by Month".

2. Membuat Bar Chart Rata-rata Delay Kedatangan:

Baris ini yang membuat visualisasi.

* sns.barplot: Fungsi ini dari Seaborn digunakan untuk membuat bar chart.
    
* x=flight_data.index: Argumen x menentukan sumbu x dari chart. Kita menggunakan flight_data.index untuk menjadikan kolom index (dalam hal ini "Month") sebagai sumbu x. Ini berarti setiap bar pada chart akan mewakili bulan tertentu.
    
* y=flight_data['NK']: Argumen y menentukan sumbu y dari chart. Kita menggunakan flight_data['NK'] untuk mengambil data delay penerbangan Spirit Airlines (kolom 'NK') sebagai nilai pada sumbu y. Dengan demikian, tinggi setiap bar akan menunjukkan rata-rata delay kedatangan untuk penerbangan Spirit Airlines di bulan tersebut.

Menambahkan Label untuk Sumbu Y:
    Baris ini menambahkan label "Arrival delay (in minutes)" ke sumbu vertikal (y) untuk menjelaskan arti dari nilai pada sumbu tersebut.

Setelah dieksekusi, kode ini akan menghasilkan bar chart yang menampilkan bar untuk setiap bulan. Tinggi setiap bar menunjukkan rata-rata delay kedatangan bulanan untuk penerbangan Spirit Airlines.

## Heatmap

Kode yang diberikan berikut ini akan menghasilkan visualisasi berupa heatmap yang menunjukkan rata-rata delay kedatangan bulanan untuk setiap maskapai penerbangan yang ada pada data Anda:

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

1. Mengatur Ukuran dan Judul (sama seperti sebelumnya):

Baris-baris awal mengatur ukuran plot menjadi 14 x 7 inch dan menambahkan judul "Average Arrival Delay for Each Airline, by Month".

2. Membuat Heatmap Delay Kedatangan Rata-rata:

Baris ini yang membuat visualisasi heatmap.

* sns.heatmap: Fungsi ini dari Seaborn digunakan untuk membuat heatmap.

* data=flight_data: Argumen data menentukan data yang akan divisualisasikan sebagai heatmap. Dalam kasus ini, kita menggunakan seluruh DataFrame flight_data yang berisi informasi delay penerbangan per bulan untuk berbagai maskapai.

* annot=True: Argumen annot=True menambahkan nilai rata-rata delay pada setiap kotak kecil (cell) pada heatmap. Ini membantu pembaca memahami nilai delay rata-rata untuk setiap kombinasi maskapai dan bulan.

3. Menambahkan Label untuk Sumbu X:
    Baris ini menambahkan label "Airline" ke sumbu horizontal (x) untuk menjelaskan arti dari sumbu tersebut.

Setelah dieksekusi, kode ini akan menghasilkan heatmap di mana sumbu x mewakili maskapai penerbangan (berdasarkan kode maskapai) dan sumbu y mewakili bulan. Warna pada setiap kotak kecil menunjukkan rata-rata delay kedatangan untuk maskapai tersebut di bulan tertentu. Nilai rata-rata delay juga akan ditampilkan di dalam kotak kecil tersebut.
